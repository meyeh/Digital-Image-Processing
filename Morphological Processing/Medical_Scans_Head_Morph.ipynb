## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from skimage import io, exposure, filters, color
import cv2
```

## Import Image

```python
image = io.imread("/content/HW2_Head.tif")

# Check Shape
print(f"Original Shape: {image.shape}")

# Plot Image
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Dilation

```python
# Structing Element
structure_element = np.ones((3,3))

# Apply SE to Image
dilated_image = cv2.dilate(image, structure_element)

# Plot
plt.imshow(dilated_image, cmap='gray')
plt.title("Dilated Image")
plt.axis('off')
plt.show()
```

## Erosion

```python
# Apply SE to Image
eroded_image = cv2.erode(image, structure_element)

# Plot
plt.imshow(eroded_image, cmap='gray')
plt.title("Eroded Image")
plt.axis('off')
plt.show()
```

# Morphological Gradient

```python
# Gradient equation
gradient = dilated_image - eroded_image

# Plot
plt.imshow(gradient, cmap='gray')
plt.title("Gradient Image")
plt.axis('off')
plt.show()
```

## Final Images

```python
plt.figure(figsize=(20,20))

plt.subplot(1,4,1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(dilated_image, cmap="gray")
plt.title("Dilated Image")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(eroded_image, cmap="gray")
plt.title("Eroded Image")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(gradient, cmap="gray")
plt.title("Gradient Image")
plt.axis("off")

plt.show()
```
