## Grey Level Transformation

```python
import matplotlib.pyplot as plt
from skimage import io, color, exposure
import numpy as np
```

```python
# Load an image
image_path = "C:/Users/megan/OneDrive/Desktop/DIP Images/heq.jpg"
image = io.imread(image_path)  # read the image by using imread
print(image.shape)  # check the shape. If 1 then it is grayscale, if it is 3 then it is color image. 
```

```python
# Convert image to greyscale if it is in color
if len(image.shape) == 3:   # If it is 3 (color), then convert it to grayscale image to get one matrix. Then mulitply by 255 to scale it.
    image = color.rgb2gray(image)
image = image * 255
plt.imshow(image, cmap = "gray")  # Use imshow to show the image. Without cmap, it will egnerat zudo color?
```

# Image Negative

```python
import numpy as np
```

```python
np.max(image)
im_negative = np.max(image) - image
plt.imshow(im_negative, cmap = "gray")
```

## Histogram

```python
table_image = io.imread("C:/Users/megan/OneDrive/Desktop/DIP Images/table.jpg")

#Convert to gray
if len(table_image.shape) == 3: 
    image = color.rgb2gray(table_image)
image = image * 255
plt.imshow(image, cmap = "gray")

plt.show()
```

```python
hist, bins_center = exposure.histogram(table_image, nbins = 256)

plt.imshow(image, cmap = "gray")
plt.show()

plt.bar(bins_center, hist, width = 0.1)
plt.show()
```

## Histogram Equalization

```python
image_eq = exposure.equalize_hist(image)
hist2, bins_center = exposure.histogram(image_eq, nbins = 256)
plt.bar(bins_center * 255, hist2, width = 1)
plt.show()
plt.imshow(image_eq, cmap ="gray")
```
